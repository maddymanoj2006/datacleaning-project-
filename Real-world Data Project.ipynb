```python
# ============================================
# REAL-WORLD DATA PROJECT
# RETAIL SALES ANALYSIS
# ============================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option('display.max_columns', None)

# ============================================
# Load Dataset
# ============================================

df = pd.read_csv("/content/mobile_sales.csv")

# ============================================
# DATA UNDERSTANDING
# ============================================

print("="*50)
print("FIRST 5 ROWS")
print("="*50)
print(df.head())

print("\n")

print("="*50)
print("DATASET SHAPE")
print("="*50)
print(df.shape)

print("\n")

print("="*50)
print("COLUMN NAMES")
print("="*50)
print(df.columns)

print("\n")

print("="*50)
print("DATASET INFO")
print("="*50)
print(df.info())

print("\n")

# ============================================
# DATA CLEANING
# ============================================

print("="*50)
print("MISSING VALUES")
print("="*50)
print(df.isnull().sum())

print("\n")

print("="*50)
print("DUPLICATE ROWS")
print("="*50)
print(df.duplicated().sum())

print("\n")

# ============================================
# STATISTICAL SUMMARY
# ============================================

print("="*50)
print("STATISTICAL SUMMARY")
print("="*50)
print(df.describe())

# ============================================
# CUSTOMER GENDER DISTRIBUTION
# ============================================

plt.figure(figsize=(7,5))

sns.countplot(
    x='CustomerGender',
    data=df
)

plt.title("Customer Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()

# ============================================
# PAYMENT METHOD ANALYSIS
# ============================================

plt.figure(figsize=(8,5))

sns.countplot(
    x='PaymentMethod',
    data=df
)

plt.title("Payment Method Distribution")
plt.xlabel("Payment Method")
plt.ylabel("Count")

plt.xticks(rotation=20)

plt.show()

# ============================================
# CUSTOMER AGE DISTRIBUTION
# ============================================

plt.figure(figsize=(8,5))

sns.histplot(
    df['CustomerAge'],
    bins=10,
    kde=True
)

plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()

# ============================================
# PRICE DISTRIBUTION
# ============================================

plt.figure(figsize=(8,5))

sns.histplot(
    df['Price'],
    bins=15,
    kde=True
)

plt.title("Mobile Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()

# ============================================
# REVENUE DISTRIBUTION
# ============================================

plt.figure(figsize=(8,5))

sns.histplot(
    df['TotalRevenue'],
    bins=20,
    kde=True
)

plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")

plt.show()

# ============================================
# TOP 10 MOBILE MODELS BY REVENUE
# ============================================

top_models = (
    df.groupby('MobileModel')['TotalRevenue']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))

top_models.plot(
    kind='bar'
)

plt.title("Top 10 Mobile Models by Revenue")
plt.xlabel("Mobile Model")
plt.ylabel("Revenue")

plt.show()

# ============================================
# TOP 10 BRANDS BY REVENUE
# ============================================

top_brands = (
    df.groupby('Brand')['TotalRevenue']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))

top_brands.plot(
    kind='bar'
)

plt.title("Top 10 Brands by Revenue")
plt.xlabel("Brand")
plt.ylabel("Revenue")

plt.show()

# ============================================
# PRICE VS REVENUE
# ============================================

plt.figure(figsize=(8,5))

sns.scatterplot(
    x='Price',
    y='TotalRevenue',
    data=df
)

plt.title("Price vs Revenue")

plt.show()

# ============================================
# UNITS SOLD VS REVENUE
# ============================================

plt.figure(figsize=(8,5))

sns.scatterplot(
    x='UnitsSold',
    y='TotalRevenue',
    data=df
)

plt.title("Units Sold vs Revenue")

plt.show()

# ============================================
# REVENUE BY PAYMENT METHOD
# ============================================

plt.figure(figsize=(8,5))

sns.boxplot(
    x='PaymentMethod',
    y='TotalRevenue',
    data=df
)

plt.title("Revenue by Payment Method")

plt.xticks(rotation=20)

plt.show()

# ============================================
# CORRELATION HEATMAP
# ============================================

numeric_cols = [
    'Price',
    'UnitsSold',
    'TotalRevenue',
    'CustomerAge'
]

corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title("Correlation Heatmap")

plt.show()

# ============================================
# LOCATION ANALYSIS
# ============================================

top_locations = (
    df['Location']
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_locations.index,
    y=top_locations.values
)

plt.title("Top 10 Locations")
plt.xticks(rotation=45)

plt.show()

# ============================================
# BUSINESS INSIGHTS
# ============================================

print("\n")
print("="*50)
print("BUSINESS INSIGHTS")
print("="*50)

print("\nMost Used Payment Method:")
print(df['PaymentMethod'].value_counts().idxmax())

print("\nMost Popular Brand:")
print(df['Brand'].value_counts().idxmax())

print("\nMost Sold Mobile Model:")
print(df['MobileModel'].value_counts().idxmax())

print("\nAverage Mobile Price:")
print(round(df['Price'].mean(), 2))

print("\nAverage Revenue:")
print(round(df['TotalRevenue'].mean(), 2))

print("\nAverage Customer Age:")
print(round(df['CustomerAge'].mean(), 2))

print("\nHighest Revenue Brand:")
print(
    df.groupby('Brand')['TotalRevenue']
    .sum()
    .idxmax()
)

print("\nHighest Revenue Mobile Model:")
print(
    df.groupby('MobileModel')['TotalRevenue']
    .sum()
    .idxmax()
)

print("\nTop Location:")
print(df['Location'].value_counts().idxmax())

print("\nProject Completed Successfully!")
```
